# **01-1 파이썬을 하기 전에 읽는 아주 간단한 설명**

**컴퓨터 프로그램이란**

 '컴퓨터가 무엇을 해야 할지 미리 작성한 진행 계획'입니다. 예를 들어 카카오톡에서 첫 번째로 친구를 지정하고 다음에 메시지를 입력하고 마지막으로 전송 버튼을 누르면 메시지가 전달되는 형태이다. 이는 카카오톡이라는 프로그램이 카톡을 보낼 사람을 지정하고, 카톡을 입력하고, 전송 버튼을 누르는 프로그램이 작성되어 있기 때문이다. 

**프로그래밍 언어**

1. 공책의 오른쪽 위에 a라고 적습니다.
2. a 옆에 10이라고 적습니다.
3. 공책의 왼쪽 위에 b라고 적습니다.
4. b 옆에 20이라고 적습니다.
5. 공책의 오른쪽 아래에 c라고 적습니다.
6. a 옆에 있는 숫자와 b 옆에 있는 숫자를 더해서 c 옆에 적습니다.
7. 만약 c 옆에 적혀 있는 숫자가 15보다 크면 일어나서 "15보다 크다!"라고 외칩니다.
8. 만약 c엎에 적혀 있는 숫자가 15보다 작으면 일어나서 "15보다 작다!"라고 외칩니다.
이진 숫자: 0과 1로 이루어진 것으로 컴퓨터가 유일하게 이해할 수 있는 값
이진 코드: 명령어를 컴퓨터가 이해하기 위한 저급 언어로 바꾸면 이진 코드가 된다.
소스 코드: 프로그래밍 언어로 작성한 코드
프로그래밍 언어: 사람이 이해하기 쉽게 프로그램을 만들기 위해서 만들어진 언어
코드 실행기: 컴퓨터가 이해하기 힘든 프로그래밍 언어를 컴퓨터가 알아듣기 쉽게 번역하는 일종의 번역기

**프로그래밍 언어, 파이썬**

장점
1. 문접이 상대적으로 단순해서 쉽게 배울 수 있다.
2. 다양한 분야에 활용할 수 있다.
3. 대부분의 운영체제에서 동일하게 사용하기 때문에 범용성이 높다.
단점
1. c언어보다 10~350배 정도 느리다.
