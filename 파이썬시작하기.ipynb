**키워드**

*주의점: 식별자가 키워드가 같으면 안 된다.

image.png

파이썬은 대소문자를 구분합니다. 예를 들어서 True는 키워드지만 true는 키워드가 아닙니다. True로는 이름을 정할 수 없고 true로는 이름을 정할 수 있다.

이후에 사용되는 단어가 키워드인지 아닌지 확인해야 할 경우가 있을 수 있습니다.

```python
import keyword
print(keyword.kwlist)
```

**식별자**
시별자는 프로그래밍 언어에서 이름을 붙일 때 사용되는 단어입니다. 주로 벼누나 함수 이름 등으로 사용됩니다.

식별자의 4가지 규칙

1. 키워드를 사용하면 안 됩니다.
2. 특수 문자는 언더 바(_)만 허용됩니다.
3. 숫자로 시작하면 안 됩니다.
4. 공백을 포함할 수 없습니다.

위 규칙에 맞는 단어는 모두 식별자로 사용할 수 있습니다. 다음 표에서 왼쪽의 단어는 모두 식별자로 사용할 수 있지만, 오른쪽의 단어는 모두 식별자로 사용할 수 없습니다. 식별자로 사용할 수 없는 이유는 중요하다!

사용가능한 단어

alpha

alpha10

_alpha

Alpha

ALPHA

사용 불가능한 단어

break, 이유: 키워드라서 안 됩니다.

273alpha, 이유: 숫자로 시작해서는 안 됩니다.

has space, 이유: 공백을 포함해서 안 됩니다.

식별자로 만들 때는 한글, 한자, 일본어와 같은 전 세계의 언더를 모두 사용할 수 있지만, 알파벳을 사용하는 것이 관례입니다. 또한 a,b처럼 의미 없는 단어보다 file, ouput처럼 의미 있는 단어를 사용하는 것이 좋습니다.

스네이크 케이스와 캐멀 케이스

식별자에는 공백을 사용할 수 없습니다. 다음과 같은 식별자의 의미를 쉽게 이해할 수 있나요?

itemlist, loginstatus, characterhp, rotateangle

위의 경우 이해는 가지만, 공백이 없어서 재빠르게 이해하기 힘들가 그래서 스네이크 케이스(snake_case)와 캐멀 케이스(CamelCase) 두 가지를 사용합니다.

첫째, 단어 사이에 언더 바(_) 기호를 붙여 식별자를 만든다.

예: item_list

둘째, 단어들의 첫 글자를 대문자로 만들어 식별자를 만듭니다.

예: ItemList

일반 식별자

itemlist, loginstatus, characterhp, rotateangle

스네이크 케이스

item_list, login_status, character_hp, rotate_angle

케멀 케이스

ItemList, LoginStatus, CharacterHp, RotateAngle

**식별자 구분하기**

원래 케멀 케이스는 '첫 번째 글자를 대문자로 적는다'와 '첫 번째 글자를 소문자로 적는다'로 구분합니다. 하지만 파이썬에서는 '첫 번째 글자를 소문자로 적는다'라는 캐멀 케이스는 사용하지 않습니다.

캐멀 케이스 유형1: PrintHello, 파이썬에서 사용

캐멀 케이스 유형2: printHello, 파이썬에서 사용하지 않습니다.

따라서 첫 글자가 소문자이면 스네이크

print, input, list, str, map, filter

반대로 첫 글자가 대문자이면 무조건 캐멀 케이스입니다.

Animal, Customer
